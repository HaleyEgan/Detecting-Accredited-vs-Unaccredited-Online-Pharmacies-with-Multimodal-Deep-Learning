```python
### TEST
#url = 'https://acariahealth.envolvehealth.com' #'https://www.carolinacarerx.org'
#image_url = 'https://www.carolinacarerx.org'

def url_to_df(url):
    try:
        html = urlopen(url)
        headers = {'user-agent': 'Mozilla/5.0'}
        page = requests.get(url, headers)
        soup = BeautifulSoup(page.content,'html.parser')
        body = soup.body
        text = []
        for s in body.strings:
            text.append(s)
    
        images = []    
        for item in soup.find_all('img'):
            images.append(url + item['src'])
    
        image2 = []
        for item in soup.find_all('img'):
            image2.append(item['src'])
        url_df = pd.DataFrame(data= {'url':[url],'text':[text],
                                     'image_url':[images], 'image_source':[image2]})
        return url_df
    except: 
        pass
```

```python
df = pd.DataFrame()

for u in url_list:
    dictObj =url_to_df(u)
    df = df.append(dictObj) # can we suppress/ignore the deprecation messages ??
```
